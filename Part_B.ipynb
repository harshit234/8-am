```python
# tax_calculator.py

income = float(input("Enter annual income (₹): "))

STANDARD_DEDUCTION = 75000
taxable_income = max(0, income - STANDARD_DEDUCTION)

slabs = [
    (300000, 0.00),
    (700000, 0.05),
    (1000000, 0.10),
    (1200000, 0.15),
    (1500000, 0.20),
    (float('inf'), 0.30)
]

previous_limit = 0
tax = 0

print("\nTax Breakdown:")

for limit, rate in slabs:

    if taxable_income > previous_limit:
        taxable_amount = min(taxable_income, limit) - previous_limit
        slab_tax = taxable_amount * rate

        print(f"Income {previous_limit} - {limit}: ₹{taxable_amount} taxed at {rate*100}% = ₹{slab_tax}")

        tax += slab_tax
        previous_limit = limit

    else:
        break

effective_rate = (tax / income) * 100 if income > 0 else 0

print("\nTotal Tax:", tax)
print("Effective Tax Rate:", round(effective_rate, 2), "%")
```
