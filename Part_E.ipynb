PART E — Smart Transaction Validator

```python
# transaction_validator.py

amount = float(input('Transaction amount (Rs): '))
category = input('Category (food/travel/electronics/other): ').lower()
hour = int(input('Hour of transaction (0-23): '))
daily_spent = float(input('Amount already spent today (Rs): '))
vip = input("VIP customer? (yes/no): ").lower() == "yes"

# Limits
single_limit = 50000 * (2 if vip else 1)
daily_limit = 100000 * (2 if vip else 1)

category_limits = {
    "food": 5000 * (2 if vip else 1),
    "electronics": 30000 * (2 if vip else 1),
    "travel": 50000 * (2 if vip else 1),
    "other": 20000 * (2 if vip else 1)
}

# BLOCK RULES
if amount > single_limit:
    print("BLOCKED: exceeds single transaction limit")

elif daily_spent + amount > daily_limit:
    print("BLOCKED: exceeds daily spending limit")

# FLAG RULES
elif hour < 6 or hour > 23:
    print("FLAGGED: unusual transaction time")

elif amount > category_limits.get(category, 20000):
    print("FLAGGED: exceeds category limit")

else:
    print("APPROVED")
```
