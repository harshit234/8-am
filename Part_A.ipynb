```python
# student_admission_system.py

def validate_inputs(entrance_score, gpa, recommendation, category, extracurricular):
    if not (0 <= entrance_score <= 100):
        return "Invalid entrance score (0-100 allowed)"

    if not (0 <= gpa <= 10):
        return "Invalid GPA (0-10 allowed)"

    if recommendation not in ["yes", "no"]:
        return "Recommendation must be 'yes' or 'no'"

    if category not in ["general", "obc", "sc_st"]:
        return "Invalid category"

    if not (0 <= extracurricular <= 10):
        return "Extracurricular score must be 0-10"

    return None


def admission_decision(entrance_score, gpa, recommendation, category, extracurricular):

    error = validate_inputs(entrance_score, gpa, recommendation, category, extracurricular)
    if error:
        return f"REJECTED: {error}"

    # Merit rule
    if entrance_score >= 95:
        return "ADMITTED (Scholarship): Exceptional entrance score"

    bonus = 0
    bonus_details = []

    if recommendation == "yes":
        bonus += 5
        bonus_details.append("+5 recommendation")

    if extracurricular > 8:
        bonus += 3
        bonus_details.append("+3 extracurricular")

    effective_score = entrance_score + bonus

    cutoff = {
        "general": 75,
        "obc": 65,
        "sc_st": 55
    }

    if gpa < 7:
        return "REJECTED: GPA below minimum requirement"

    if effective_score >= cutoff[category]:
        return f"""
ADMITTED (Regular)

Bonus Applied: {' '.join(bonus_details)}
Effective Score: {effective_score}

Reason: Meets {category.upper()} cutoff ({effective_score} ≥ {cutoff[category]}) and GPA requirement ({gpa} ≥ 7)
"""

    elif effective_score >= cutoff[category] - 5:
        return "WAITLISTED: Slightly below cutoff"

    else:
        return "REJECTED: Entrance score below cutoff"


# Input section
entrance = int(input("Entrance Score: "))
gpa = float(input("GPA: "))
rec = input("Recommendation (yes/no): ").lower()
category = input("Category (general/obc/sc_st): ").lower()
extra = float(input("Extracurricular Score: "))

result = admission_decision(entrance, gpa, rec, category, extra)
print(result)
```
